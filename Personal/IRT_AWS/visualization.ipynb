```python
import matplotlib.pyplot as plt
import numpy as np

# 예시 데이터
item_index = list(range(1, 15))  # 문항 번호
estimated_theta = np.random.normal(0, 0.5, size=14)  # 추정된 θ 값 (예시)
se_theta = 1 / np.sqrt(np.arange(1, 15))  # 표준오차(SE(θ))

# 시각화
plt.figure(figsize=(10, 6))
plt.errorbar(
    item_index,
    estimated_theta,
    yerr=se_theta,
    fmt="o",
    color="purple",
    ecolor="red",
    capsize=5,
)
plt.axhline(y=0, color="gray", linestyle="--", linewidth=0.5)
plt.xlabel("Item Index")
plt.ylabel("Estimated Ability (Theta)")
plt.title("Ability Estimation with Confidence Intervals")
plt.grid(True)
plt.show()
```
