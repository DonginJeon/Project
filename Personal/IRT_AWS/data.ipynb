```python
import pandas as pd
import numpy as np

# Set random seed for reproducibility
np.random.seed(100)

# 1. Generate 30 sample question data with item parameters
question_data = {
    'question_id': [f'Q{i+1}' for i in range(30)],
    'difficulty': np.random.normal(0, 1, 30),           # Difficulty parameter (b)
    'discrimination': np.random.uniform(0.5, 2, 30),    # Discrimination parameter (a)
    'guessing': np.random.uniform(0, 0.25, 30)          # Guessing parameter (c)
}
question_df = pd.DataFrame(question_data)

# 2. Generate response data for visualization (100 students answering 30 questions)
n_students = 10
response_data = {
    'student_id': [f'ST{i+1}' for i in range(n_students)],
}

# Simulate students' abilities (theta values)
student_abilities = np.random.normal(0, 1, n_students)
response_matrix = []

# Generate responses based on 3PL model for each question
for idx, row in question_df.iterrows():
    p_correct = row['guessing'] + (1 - row['guessing']) / (
        1 + np.exp(-1.7 * row['discrimination'] * (student_abilities - row['difficulty']))
    )
    responses = np.random.binomial(1, p_correct)
    response_matrix.append(responses)

# Convert response matrix to DataFrame and transpose to match student-wise data
response_df = pd.DataFrame(response_matrix).T
response_df.columns = [f'Q{i+1}' for i in range(30)]
response_df['student_id'] = response_data['student_id']
response_df.set_index('student_id', inplace=True)

# Save to CSV
question_df.to_csv("question_data.csv", index=False)
response_df.to_csv("response_data.csv")
```

```python
import pandas as pd
import numpy as np

# 문제 수 설정
num_questions = 30

# 1PL 문제 데이터 생성 (난이도 b 값만 포함)
data_1PL = {
    'question_id': [f'Q{i+1}' for i in range(num_questions)],
    'difficulty_b': np.random.uniform(-2, 2, num_questions)  # 난이도 b 값은 -2에서 2 사이 랜덤 값
}
df_1PL = pd.DataFrame(data_1PL)
df_1PL.to_csv('question_data_1PL.csv', index=False)

# 2PL 문제 데이터 생성 (난이도 b 값과 변별도 a 값 포함)
data_2PL = {
    'question_id': [f'Q{i+1}' for i in range(num_questions)],
    'difficulty_b': np.random.uniform(-2, 2, num_questions),  # 난이도 b 값은 -2에서 2 사이 랜덤 값
    'discrimination_a': np.random.uniform(0.5, 2.0, num_questions)  # 변별도 a 값은 0.5에서 2.0 사이 랜덤 값
}
df_2PL = pd.DataFrame(data_2PL)
df_2PL.to_csv('question_data_2PL.csv', index=False)

# 3PL 문제 데이터 생성 (난이도 b 값, 변별도 a 값, 추측도 c 값 포함)
data_3PL = {
    'question_id': [f'Q{i+1}' for i in range(num_questions)],
    'difficulty_b': np.random.uniform(-2, 2, num_questions),  # 난이도 b 값은 -2에서 2 사이 랜덤 값
    'discrimination_a': np.random.uniform(0.5, 2.0, num_questions),  # 변별도 a 값은 0.5에서 2.0 사이 랜덤 값
    'guessing_c': np.random.uniform(0.0, 0.25, num_questions)  # 추측도 c 값은 0에서 0.25 사이 랜덤 값
}
df_3PL = pd.DataFrame(data_3PL)
df_3PL.to_csv('question_data_3PL.csv', index=False)
```
